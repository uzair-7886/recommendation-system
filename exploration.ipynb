```python
import pandas as pd
import numpy as np
```

```python
credits=pd.read_csv("tmdb_5000_credits.csv")
movie_df=pd.read_csv("tmdb_5000_movies.csv")
```

```python
credits=credits.rename(columns={"movie_id":"id"})
```

```python
movie_df=movie_df.rename(columns={"original_title":"title"})
```

```python
credits.head()
```

```python
movie_df.head()
```

```python
merged_ds=movie_df.merge(credits,on="id")
```

```python
merged_ds.head()
```

```python
columns_to_remove=["budget","genres","homepage","keywords","original_language","production_companies","runtime","overview","title_y","revenue","cast","crew","production_countries"]
merged_ds=merged_ds.drop(columns=columns_to_remove)
```

```python
df=merged_ds
df.head()
```

```python
m=300
C=df.vote_average.mean()
v=df.vote_count
R=df.vote_average

W=(R*v + C*m)/(v+m)
```

```python
df['weighted_average']=W
```

```python
df=df.sort_values('weighted_average',ascending=True)
df.head()
```

```python
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
```

```python
new_df=df[['weighted_average','popularity']]
new_df.head()
```

```python
# Plotting the graph
plt.scatter(new_df['popularity'], new_df['weighted_average'])
plt.xlabel('Popularity')
plt.ylabel('Weighted Average')
plt.title('Popularity vs Weighted Average')

```

```python
# Normalizing
scaler=MinMaxScaler()
new_df[['popularity','weighted_average']]=scaler.fit_transform(new_df)

plt.scatter(new_df['popularity'], new_df['weighted_average'])
plt.xlabel('Popularity')
plt.ylabel('Weighted Average')
plt.title("Popularity vs Weighted Average after Normalization")
plt.show()
```
